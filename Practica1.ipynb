despues de ya tener nuestra maqui virtual funcionando tuvimos queaprender a usar la consola de la maquina virtual, en lo personal tuve varios problemas con dicha consola, como problemas para instalar unos programas que necesitabamos, cosa que no me pasó con la line de comandos de la computadora(cmd) y no me quedó claro el porque pasaba eso o cuál era la diferencia por la que en una si lo podia hacer pero en otra no.

Las primeras cosas que tuvimos que hacer en esta practica fue instalar varias cosas para poder trabajar, una de ellas y con la que comenzamos fue **Nodejs**, primeramente actualizamos algunos paquetes para su uso con el comando: **sudo apt-get update**, donde **sudo** se utiliza para dar privilegios de administrador a un usuario que no lo es, una vez instalados los paquetes instalamos **Nodejs** con el comando: **sudo apt-get install nodejs** y en seguida instalaremos **NPM** el cual es un gestor de paquetes en lenguaje JavaScript, que sirve para obtener cualquier librería con tan solo una sencilla línea de código, lo cual nos permitirá agregar dependencias de forma simple más adelante, para esto utilizamos el comando: **sudo apt-get install npm**, para terminar esta primera parte necesitaremos instalar el **generador de aplicaciones Express, Express.js** este es un framework para Node.js que sirve para ayudarnos a crear aplicaciones web, para esto utilizaremos nustro comando: **sudo npm install express-generator@4 –g**.

Con esto terminamos la primera parte, en la segunda parte ue es la intalación de PostgreSQL, fué en donde más problemas tuve para poder instalarlo desde la consola de la máquina virtual y en donde tuve que aprender a instalarlo desde la línea de comando(cmd), para instalar **postgreSQL** usaremos el siguente comando: **sudo apt install postgresql postgresql-contrib**, también utilizamos unos cuantos comandos para realizar diversas acciones los cuales pondre a continuación:

1.   **Ingresar al súperusuario postgres: sudo -i -u postgres**
2.   **Abrir PostgreSQL: psql**
3.   **Salir: \q**
4.   **Crear una base desde PostgreSQL: createdb**
5.   **Crear una base de datos como súperusuario: sudo -u postgres createdb test**
6.  **Regresar al usuario: su - usuario**
7.   **Ingresar a PostgreSQL sin cambiar de usuario: sudo -u postgres psql**

Una vez instalado postgreSQL, comenzaremos con la creación del proyecto, para esto usaremos un par de comandos básicos para preparar todo el terreno, primero nos situaremos en la carpeta **home** con el comando **cd /home**, en seguida crearemos una carpeta nueva para nuestro proyecto, la cual llamaremos **practica_01** con ayuda del comando **sudo mkdir practicas cd practicas**, también es importante mencionar que necesitaremos tener abierto el puerto 3000.

Utilizando lo que instalamos al principio, inicializamos el proyecto con la ayuda de EXpress con el comando: **sudo express practica_01**, nos cambiaremos de directorio con el comando: **cd practica_01** e instalaremos otro pequeño paquete con **sudo npm install**, una vez instalado inicializaremos el servicio con: **sudo npm start**.
ahora instalaremos un par de cosas más, **pg-promise** el cual nos ayudara a evitar la cancelación de las consultas, por lo tanto nos ayudara a tener un diseño correcto de la base de datos y el otro es **Bluebird** el cual es una biblioteca que nos ayudara a mantener un buen rendimiento, para esto usaremos **sudo npm install pg-promise@5 –save, sudo npm install bluebird@3 –sabe** respectivamente.

Despúes de esto tendremos que hacer un par de cosas que parecen(y hasta cierto punto lo son) muy enrredados, pero una vez entendiendo lo que hay que hacer se vuelven muy sencillos, primeramente tenemos que crear en la raíz del proyecto un archivo, para el cual será necesario pegar las siguientes instrucciones:

El último paso para terminar la API son las rutas, para esto ingresaremos a la carpeta rutas que se encuentra en la raíz del proyecto con: **cd routes/**, una vez estando dentro modificaremos el archivo **/routes/index.js** usando el siguiente código:

var express = require('express');

var router = express.Router();

var db = require('../queries');


router.get('/api/puppies', db.getAllPuppies);

router.get('/api/puppies/:id', db.getSinglePuppy);

router.post('/api/puppies', db.createPuppy);

router.put('/api/puppies/:id', db.updatePuppy);

router.delete('/api/puppies/:id', db.removePuppy);

module.exports = router;
